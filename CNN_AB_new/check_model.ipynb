```python
import numpy as np
import re

aucs = []

with open("/work/submit/haoyun22/FCC-Beam-Background/nn.txt", "r") as f:
    for line in f:
        match = re.search(r"AUC\s*=\s*([0-9.]+)", line)
        if match:
            aucs.append(match.group(1))

print(aucs)
print(type(aucs))
```

```python
# Compute the mean AUC and standard deviation
auc_values = np.array(aucs, dtype=float)
mean_auc = np.mean(auc_values)
std_auc = np.std(auc_values)
length = len(auc_values)
print(f"Number of AUC values: {length}")
print(f"Mean AUC: {mean_auc:.5f}, Standard Deviation: {std_auc:.5f}")
```

```python
import pandas as pd

file_path = "/work/submit/haoyun22/FCC-Beam-Background/CNN_AB_new/"

# reading multiple CSV files, skip the first row and concatenating them into a single DataFrame
df_list = []

# detect all csv files in the directory
import os

for filename in os.listdir(file_path):
    if filename.endswith(".csv"):
        df = pd.read_csv(os.path.join(file_path, filename)).iloc[:, 1:]
        df_list.append(df)
combined_df = pd.concat(df_list, ignore_index=True)
combined_df.head(10)

# compute the mean and std for each column in 5 decimal places
length = combined_df.shape[0]
print(f"Number of rows in combined DataFrame: {length}")
mean_series = combined_df.mean().round(5)
std_series = combined_df.std().round(5)
result_df = pd.DataFrame({'Mean': mean_series, 'Std': std_series})
print(result_df)
```
